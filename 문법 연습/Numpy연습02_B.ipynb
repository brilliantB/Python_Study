```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = [1, 2, 3, 4, 5]
plt.plot(y)
plt.show()
```

```python
# plot() 메소드의 입력 파라미터의 개수가 2개인 경우

y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(range(len(y)),y);
```

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']   # x축에 의미있는 문자(열) 넣기
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]
plt.plot(x, y1, label='sold')  
plt.plot(x, y2, label='On Shelves')
plt.xlabel('Daily')                          # xlabel() : x축의 라벨을 입력
plt.ylabel('No. of Books Sold')              # ylabel() : y축의 라벨을 입력
plt.legend(loc="upper left")                 # 레전드(legend) 붙이기
plt.title('Le Petit Prince: Sold & Left');   # 제목 넣기
```

### Numpy를 이용한 plot기능

```python
points = np.array([[1, 1], [1, 2], [1, 3], [2, 1], [2, 2], [2, 3], [3, 1], [3, 2], [3, 3]])
p = np.array([2.5, 2])

plt.plot(points[:,0], points[:,1], "ro")
plt.plot(p[0], p[1], "bo");
```

```python
x = np.linspace(1, 10, 20)  
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=12, label="First")
plt.plot(x, y2, "gs-", linewidth=2, markersize=12, label="Second")
plt.xlabel("x")
plt.ylabel("y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```
