```python
# random 메소드에 입력값으로 5를 입력해주면 0에서 1사이의 무작위의 수 5개를 생성
# print() : 리스트의 형태로 출력
# all() : 배열의 모든 원소가 조건을 만족하면 True를 출력, 그렇지 않으면 False를 출력
# any() : 배열에 조건을 만족하는 원소가 있으면 True를 출력

ary = np.random.rand(5)
print(ary)
print(np.all(ary >= 0.3))
print(np.any(ary > 0.7))
```

### 불리언 인덱스
- 불리언 인덱스는 특정 조건을 만족하는 값을 선택(필터링)할 때 많이 사용

```python
scores = [80, 90, 70, 65, 85, 95, 90, 80, 75, 80]
score_arr = np.array(scores)
score_arr
```

```python
score_arr >= 90
```

```python
score_arr[score_arr >= 90] 
```

```python
# 90점 이상의 점수를 999점으로 변경하면 결과가 바뀜
score_arr[score_arr >= 90] = 999
score_arr
```

### 변형 (Trandspose)
- 기존에 있는 배열의 축을 바꾸는 방법으로 사용

```python
#첫번째 행(row)에 있는 값들이 첫번째 열(column)에 위치하고, 
# 두번째 행에 있는 값들이 두번째 열에 위치하도록 변환

arr = np.arange(10).reshape(2,5)
arr
```

```python
# transpose() 호출 시 축을 지정하지 않을 경우

arr.T      #arr.transpose()와 동일
```

```python
# 3차원의 축배열
arr3d = np.arange(27).reshape(3,3,3)
arr3d
```

```python
arr3d.transpose()     # transpose(2,1,0)
```

```python
arr3d.transpose(1,0,2)
```

### 제로 벡터와 제로 행렬

```python
# 0으로 초기화된 1차원의 zero_vector 생성
zero_vector = np.zeros(3)
print(zero_vector)
```

```python
# 0으로 초기화된 (4X3)형식의 2차원 배열 생성 
zero_matrix = np.zeros((4,3))
print(zero_matrix)
```

```python
# 0의 값을 갖는 5개의 1차원 배열 생성
zero_vector = np.zeros(5)
zero_vector
```

```python
# 모든 원소의 값이 0인 (5x3) 형식의 2차원 배열 생성
zero_matrix = np.zeros((5,3))
zero_matrix
```

```python
# 생성한 제로 벡터나 행렬에 새로운 값으로 다시 재설정
my_array = np.zeros(15).reshape(3, 5)
my_array += 4
my_array
```
