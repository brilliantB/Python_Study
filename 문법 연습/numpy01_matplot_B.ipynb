## 직선 그래프

```python
import matplotlib.pyplot as plt
y = [1, 2, 3, 4, 5]    # 선형
plt.plot(y)
plt.show()
```

```python
# plot() 메소드의 입력 파라미터의 개수가 2개인 경우
# (x, y)  : x축, y축
# lange(len()y) = range(7)  : 0 ~ 6까지의 리스트와 같음

y = [13, 16, 15, 18 , 16, 17, 16]
plt.plot(range(len(y)), y);
```

```python
#  X축에 의미 있는 문자열 넣기

x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(x, y);
```

```python
# 라벨(label) 입력

x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(x, y)
plt.xlabel('Daily')                             # X축의 라벨을 입력
plt.ylabel('No. of Books Sold');                # Y축의 라벨을 입력
```

```python
# 그래프 1개 더 추가

x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('Daily')                             # X축의 라벨을 입력
plt.ylabel('No. of Books Sold');                # Y축의 라벨을 입력
```

```python
# 레전드(legend) 붙이기

x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]
plt.plot(x, y1, label='sold')
plt.plot(x, y2, label='On Shelves')
plt.xlabel('Daily')                             # X축의 라벨을 입력
plt.ylabel('No. of Books Sold')                 # Y축의 라벨을 입력
plt.legend(loc="upper left");
```

```python
# 제목넣기

x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]
plt.plot(x, y1, label='sold')
plt.plot(x, y2, label='On Shelves')
plt.xlabel('Daily')                             # X축의 라벨을 입력
plt.ylabel('No. of Books Sold')                 # Y축의 라벨을 입력
plt.legend(loc="upper left");
plt.title('Le petit Prince : solad & left');
```

## 점선 그래프

```python
# 'bo' 옵션

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, 'bo')      # b(Blue) 색상의 'o' 모양 점 표시
plt.show()
```

```python
# axis() 메소드

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, 'bo')      # b(Blue) 색상의 'o' 모양 점 표시
plt.axis([0, 6, 0, 30])   # x축과 y축 범위 지정 [xmin, xmax, ymin, ymax]
plt.show()
```

## Numpy를 이용한 plot 기능

```python
import numpy as np
points = np.array([[1, 1], [1, 2], [1, 3], [2, 1], [2, 2], [2, 3], [3, 1], [3, 2], [3, 3]])
p = np.array([2.5, 2])

import matplotlib.pyplot as plt
plt.plot(points[:,0], points[:,1], "ro")
plt.plot(p[0], p[1], "bo");
```

```python
# plot() 메소드 입력 파라미터 개수에 따라 세부적인 기능이 달라짐

# import matplotlib.pyplot as plt
# import numpy as np
x = np.linspace(0, 5, 10)
y = x**2                              # ** 은 제곱을 의미
plt.plot(x, y);
```

```python
# 파라미터 개수가 다수인 경우
# 라인파란색(b), 마커모양 원(o), 선(-) ,라인폭 3, 마커사이즈 5

x = np.linspace(0, 10, 20)
y = x**2.0
plt.plot(x, y, "bo-", linewidth=3, markersize=5);   
```

```python
# 라인 연두색(g), 마커 모양 네모(s), 라인폭 1, 마커사이즈 3

plt.plot(x, y, "ys-", linewidth=100, markersize=3);
```

```python
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=12, label="First");   
plt.plot(x, y2, "ys-", linewidth=2, markersize=12, label="Second");
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```

```python
# 로그함수를 사용하여 x=np.logspace(-1, 1, 20)로 수정

x = np. logspace(-1, 1, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=5, label="First");   
plt.plot(x, y2, "gs-", linewidth=2, markersize=5, label="Second");
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.png")          # 이미지를 저장하기 (원하는 파일확장자 사용하면 저장됨, pdf, jpg, png....)
```

## 표준정규분포 그래프

```python
# standard_normal() 사용

# import matplotlib.pyplot as plt
# import numpy as np
x = np.random.standard_normal(size=1000)
plt.hist(x)
```

```python
# 결과에 빈도수가 아닌 확률로 나타내고 싶으면 hist()메소드의 인자로 density=True를 입력  (밀도)

plt.hist(x, density=True)
```

```python
# randn() 메소드의 예

x = np. random.randn(1000)
plt.hist(x)
```
