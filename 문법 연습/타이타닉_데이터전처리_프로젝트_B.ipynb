# 연령대 컬럼을 추가하시오 
컬럼이름: AgeGroup

연령대 : 미성년자, 20대, 30대, 40대, 50대, 60대이상 : 'Minor', '20s', '30s', '40s', '50s', '60+'

```python
df['AgeGroup'] = pd.cut(df.Age, bins=[10,20,30,40,50,60,100], labels=['Minor', '20s', '30s', '40s', '50s', '60+'])
df
```

# 각 연령대별 인구 수를 구하시오 

```python
df.AgeGroup.value_counts().sort_index()
```

# 각 승객의 Fare를 FareinWon이라는 원화 컬럼을 추가하시오 
- 1 USD = 1150 KRW 

```python
df['FareinWon'] = df['Fare'] * 1150
df
```

# 'Gender'라는 컬럼을 추가하여 'Sex'컬럼의 male과 female을 0과 1로 변환하시오 

```python
# sample 1

# df['Gender'] = df.Sex.astype('category',).cat.codes
# df.head()
```

```python
# sample 2

df['Gender'] = df['Sex'].map({'female' : 1, 'male': 0}).astype(int)
df.head()
```

# 객실 등급별로 인구 수를 구하시오 

```python
df.Pclass.value_counts().sort_index()
```

# 승객들의 이름에서 타이틀(Mr, Mrs, Miss, Ms., Rev. 등 )를 제거하시오. 
- 정규식 사용 


```python
df.Name.str.extract('^(\w+\.\W)')
```

```python
df.Name.str.replace('^(\w+\.\W)','')
```

```python
df['Name_NoTitle'] = df.Name.str.replace('^(\w+\.\W)','')
df.head()
```

- lamda 사용

```python
df['Name_NoTitle'] = df['Name'].apply(lambda x : x.split('.')[1])
df['Name_NoTitle']
df
```
