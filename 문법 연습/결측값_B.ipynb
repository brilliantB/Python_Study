# 결측값 제거 (dropna)
- 비어있는 값을 제거하기

- dropna 함수 사용: 변수.dropna 
-- 행의 결측값 제거: 변수.dropna(axis=0) 
-- 열의 결측값 제거: 변수.dropna(axis=1) 
-- how옵션: 'any'- 한개라도 있는 경우 제거,'all'- 모두 NaN인 경우 제거 
-변수.dropna(axis=1, how='any') : 한개라도 있는 경우 행 제거(drop) 
-변수.dropna(axis=1, how='all') : NaN 값 모두 제거(drop)

```python
# 변수 df의 데이터 프레임 중 결측값(NaN) 제거

df.dropna()
```

## dropna 옵션 : 행제거

```python
# 변수 df의 데이터 프레임 중 행 기준 결측값(NaN) 제거

df.dropna(axis=0)
```

## dropna 옵션: 열제거

```python
# 변수 df의 데이터 프레임 중 열 기준 결측값(NaN) 제거

df.dropna(axis=1)
```

## how 옵션

- how='any': 한개라도 있는 경우 제거 
- how='all': 모두 NaN인 경우 제거 
*axis=0: 행 
*axis=1: 열

```python
# 변수 df의 데이터 프레임 중 행 기준으로 한개라도 결측값(NaN)이 있으면 행 제거

df.dropna(axis=0, how='any')
```

```python
# 변수 df의 데이터 프레임 중 행 기준으로 전부 결측값(NaN)이 있으면 행 제거

df.dropna(axis=0, how='all')
```

# 데이터전처리 : Merge연습
- merge : 데이터 프레임 합치기

```python
# merge 기초

cust = pd.DataFrame({
    '고객번호' : [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름' : ['가', '나', '다', '라', '마', '바', '사']
}, columns = ['고객번호', '이름'])
cust
```

```python
order = pd.DataFrame({
    '고객번호' : [1001, 1001, 1005, 1006, 1009, 1001],
    '금액' : [10000, 20000, 15000, 5000, 100000, 30000]
}, columns = ['고객번호', '금액'])
order
```

## 내부 조인 (inner join)
- 위 두 테이블을 합치려 할 때 , 딱 봐도 고객번호 를 가지고 연결 
- SQL에서 join 과 동일한 원리이다

- 교집합같은 개념

```python
# pd.merge()를 이용해서 두 테이블을 합치기

pd.merge(cust,order)
```

## 외부 조인(outer join)
- 일단 조인 할 두 테이블의 데이터를 모두 가져오고 싶을 땐 외부 조인 (outer 을 활용한다

- 합집합같은 개념

```python
pd.merge(cust, order, how = 'outer')
```

```python
# left join

pd.merge(cust, order, how = 'left')
```

```python
# right join

pd.merge(cust, order, how = 'right')
```

## Cross join

```python
df1 = pd.DataFrame({
    '품종' : ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이' : [1.4, 1.3, 1.5, 1.3]}, 
    columns= ['품종', '꽃잎길이'])
df1
```

```python
df2 = pd.DataFrame({
    '품종' : ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎너비' : [0.4, 0.3, 0.5, 0.3]}, 
    columns= ['품종', '꽃잎너비'])
df2
```

```python
pd.merge(df1, df2)
```

## on을 사용한 merge

```python
df1 = pd.DataFrame({
    '고객명' : ['춘향', '춘향', '몽룡'],
    '날짜' : ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터' : ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명' : ['춘향', '몽룡'],
    '데이터' : ['여자','남자']})
df2
```

```python
pd.merge(df1, df2, on = '고객명')
```

## 기준이 되는 Key의 칼럼명이 다를때 (left_on, right_on 사용)

```python
df1 = pd.DataFrame({
    '이름' : ['영희', '철수', '철수'],
    '성적' : [1, 2, 3]})
df1
```

```python
# df1의 이름 ' 칼럼과 df2 의 성명 ' 칼럼을 Key 로 하여 조인을 할 경우엔 left_on, right_on 인수를 사용

df2 = pd.DataFrame({
    '성명' : ['영희', '영희', '철수'],
    '성적2' : [4, 5, 6]})
df2
```

```python
pd.merge(df1, df2, left_on='이름', right_on='성명')
```

```python
# 왼쪽에 해당하는 데이터프레임의 칼럼명을 left_on = ' " 에 적어주고 오른쪽도 같은 방식으로 적어주면 된다

df3 = pd.merge(df1, df2, left_on='이름', right_on='성명')
df3.loc[:, ["이름", "성적", "성적2"]]
```

```python
df3 = pd.merge(df1, df2, left_on='이름', right_on='성명')
df3 = df3.drop('성명', axis=1)
df3
```
