```python
import random
import tkinter as tk
import tkinter.messagebox as tmsg

# 처음에 랜덤인 4개의 숫자를 작성해 둔다
a = [random.randint(0, 9),
     random.randint(0, 9),
     random.randint(0, 9),
     random.randint(0, 9)]

# 버튼이 클릭되었을 때의 처리
def ButtonClick():
    # 텍스트 입력칸에 입력된 문자열을 취득
    b = editbox1.get()
    
    # Lesson 5~4의 프로그램으로부터 판정 부분을 빌려씀
    # 4자리수의 숫자인지를 판정한다
    isok = False
    if len(b) != 4: 
        # 메세지로서 표시한다
        tmsg.showinfo("오류", "4자리수의 숫자를 입력하세요")
    else:
        numberok = True
        for i in range(4):
            if (b[i] <"0") or (b[i] > "9") :
                tmsg.showerror("오류", "숫자가 아닙니다")
                numberok = False
                break
        if numberok :
            isok = True
            
    if isok :
        # 4자릿수의 숫자였을 때
        # 히트를 판정
        hit = 0
        for i in range(4) :
            if a[i] == int(b[i]):
                hit = hit + 1
                
        #블로를 판정
        blow = 0
        for i in range(4):
            for i in range(4):
                if (int(b[i]) == a[i]) and (a[i]) != int(b[i]) != int(b[i]) :
                    blow = blow + 1
                    break
                    
        # 히트가 4라면 맞춰서 종료
        if hit == 4:
            tmsg.showinfo("맞췄다", "축하합니다. 맞췄습니다")
            # 종료
            root.destroy()
        else :
            #히트수와 블로수를 표시
            tmsg.showinfo("힌트", "히트" + str(hit) + "/" + "블로" + str(blow) + "\n")

# 메인 프로그램

# 윈도우를 만든다
root = tk.Tk()
root.geometry("400x150")
root.title("수 맞추기 게임")

# 라벨을 만든다
label1 = tk.Label(root, text="숫자를 입력하세요", font=("Helvetica", 14))
label1.place(x = 20, y = 20)

# 텍스트 박스를 만든다
editbox1 = tk.Entry(width = 4, font=("Helvetica", 20))
editbox1.place(x = 140, y = 60)

# 버튼을 만든다
button1 = tk.Button(root, text = "확인", font=("Helvetica", 14), command=ButtonClick)
button1.place(x = 200, y = 60)

# 윈도우를 표시한다
root.mainloop()
```
