# 포털 사이트에서 기사 크롤링하기

```python
# 크롤링을 위한 기본 셋팅 - 필요한 모듈 / lib import
import os, re
import urllib.request as ur
from bs4 import BeautifulSoup as bs
```

```python
url='https://media.daum.net/'
```

```python
# 마법의 명령어 - 암기
# 결과값 저장할 객체 이름 soup(이름 정하는건 자유), 위의 url을 불러오기 위해서 urlopen에 사용
soup = bs(ur.urlopen(url).read(), "html.parser")
```

```python
#html 내용을 출력
soup
```

```python
# 특정 클래스 속성을 출력하기 : 가장 직관적인 방법
print(soup.find_all('div', {"class" : "item_issue"}))
```

```python
# a 태그 5개만 출력해보기
print(soup.find_all('a')[:5])
```

```python
# a 태그의 링크 5개만 출력해보기
for i in soup.find_all('a')[:5]:
    print(i.get('href'))
```
